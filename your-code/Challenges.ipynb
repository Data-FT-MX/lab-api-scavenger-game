```python
list_forks = list(repo.get_forks())
list_forks
```

```python
list_lang = []
for fork in list_forks:
    list_lang.append(fork.get_languages())
list_lang
```

```python
keys_list_lang = []
for el in list_lang:
    keys_list_lang.append(el.keys())
keys_list_lang
```

```python
langs = []
for el in keys_list_lang:
    langs.append(list(el))
langs
```

```python
lan_res=[]
for vector in langs:
    for el in vector:
        if el not in lan_res:
            lan_res.append(el)
lan_res
```
